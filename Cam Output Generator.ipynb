```python
# normalize shape to circle with radius of scale
max_radius = max([np.sqrt(x**2 + y**2) for x, y in projection_shape])
projection_shape = [(x / max_radius * scale, y / max_radius * scale) for x, y in projection_shape]

#find center of shape
center = np.mean(projection_shape, axis=0)

#move center to origin
projection_shape = [(x - center[0], y - center[1]) for x, y in projection_shape]
```

```python
# Create intermediate points for the projection
full_projection_shape = []

for i in range(1, len(projection_shape)):
    start_point = projection_shape[i - 1]
    end_point = projection_shape[i]
    length = np.sqrt((end_point[0] - start_point[0])**2 + (end_point[1] - start_point[1])**2)

    num_points = int(length / resolution)

    for j in range(1, num_points):
        x = start_point[0] + j * (end_point[0] - start_point[0]) / num_points
        y = start_point[1] + j * (end_point[1] - start_point[1]) / num_points
        full_projection_shape.append((x, y))

    #full_projection_shape.append(start_point)
```

```python
# Cam and laser specs
cam_center_diameter = 1.8 #inches
cam_center_radius = cam_center_diameter / 2
cam_cam_spacing = 1.5 #inches
laser_angle = 51.318 #degrees
vertical_height = 0.937 #in
laser_diameter = 0.6 #in
```

```python
#Transformation

angle_increment = 360 / len(full_projection_shape)

left_cam_angles = []
left_cam_radii = []

right_cam_angles = []
right_cam_radii = []

for i in range(len(full_projection_shape)):
    x = full_projection_shape[i][0]
    y = full_projection_shape[i][1]

    b_l = cam_cam_spacing / 2 + x
    b_r = cam_cam_spacing / 2 - x
    h = vertical_height + y
    l_l = np.sqrt(b_l**2 + h**2)
    l_r = np.sqrt(b_r**2 + h**2)
    cam_radius_l = l_l - laser_diameter / 2
    cam_radius_r = l_r - laser_diameter / 2

    cam_angle_l = laser_angle + i * angle_increment
    cam_angle_r = 180 - laser_angle - i * angle_increment

    left_cam_angles.append(cam_angle_l)
    left_cam_radii.append(cam_radius_l)

    right_cam_angles.append(cam_angle_r)
    right_cam_radii.append(cam_radius_r)

average_cam_radius = (np.mean(left_cam_radii) + np.mean(right_cam_radii)) / 2

left_cam_radii = [x * (cam_center_radius / average_cam_radius) for x in left_cam_radii]
right_cam_radii = [x * (cam_center_radius / average_cam_radius) for x in right_cam_radii]
```

```python
# Plot new Cam

l_cam_x = [r * np.cos(np.radians(a)) for r, a in zip(left_cam_radii, left_cam_angles)]
l_cam_y = [r * np.sin(np.radians(a)) for r, a in zip(left_cam_radii, left_cam_angles)]

l_cam = [(x, y) for x, y in zip(l_cam_x, l_cam_y)]

r_cam_x = [r * np.cos(np.radians(a)) for r, a in zip(right_cam_radii, right_cam_angles)]
r_cam_y = [r * np.sin(np.radians(a)) for r, a in zip(right_cam_radii, right_cam_angles)]

r_cam = [(x, y) for x, y in zip(r_cam_x, r_cam_y)]

r_cam_x = [x + cam_cam_spacing for x in r_cam_x]

plt.plot(l_cam_x, l_cam_y, 'go')
plt.plot(r_cam_x, r_cam_y, 'ro')
plt.axis('equal')
```

```python
# Export shapes as dxf
def points_to_svg(points, filename, scale=100):
    if not points:
        raise ValueError("The list of points is empty.")

    # Scale the points to enlarge the output
    scaled_points = [(x * scale, y * scale) for x, y in points]

    # Determine the bounding box of the scaled points to set the SVG canvas size
    min_x = min(point[0] for point in scaled_points)
    max_x = max(point[0] for point in scaled_points)
    min_y = min(point[1] for point in scaled_points)
    max_y = max(point[1] for point in scaled_points)
    
    width = max_x - min_x
    height = max_y - min_y

    # Create an SVG drawing
    dwg = svgwrite.Drawing(filename, profile='tiny', size=(width, height))

    # Create a polyline (a series of connected lines)
    # Invert y-coordinates to correct the vertical orientation
    polyline_points = [(x - min_x, max_y - y) for x, y in scaled_points]
    
    # Connect the last point to the first point
    polyline_points.append(polyline_points[0])
    
    polyline = dwg.polyline(points=polyline_points, stroke='black', fill='none', stroke_width=2)

    # Add the polyline to the drawing
    dwg.add(polyline)

    # Add a circle centered at (0,0) with a diameter of 0.25
    center_x = -min_x
    center_y = max_y
    radius = (0.25 / 2) * scale
    
    # Add the circle to the drawing
    circle = dwg.circle(center=(center_x, center_y), r=radius, stroke='red', fill='none', stroke_width=1)
    dwg.add(circle)

    # Save the SVG file
    dwg.save()

#Generate name with hour and minute timestamp
now = datetime.datetime.now()
timestamp = now.strftime("%H%M")


points_to_svg(l_cam, 'cam_' + timestamp + '_left.svg', scale = 96)
points_to_svg(r_cam, 'cam_' + timestamp + '_right.svg', scale= 96)
```
